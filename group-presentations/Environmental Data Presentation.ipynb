#### Environmental data is that which is based on the measurement of environmental pressures, the state of the environment and the impacts on ecosystems. This is usually the "P", "S" and "I" of the DPSIR model where D = Drivers, P = Pressures, S = State, I = Impact, R = Response. Environmental data is typically generated by institutions executing environmental law or doing environmental research .

![image](http://aqma.defra.gov.uk/images/aqma_maps/Sheffield.jpg)

# Different types of environmental data

## 1.Temperature and wind

### Temperature and wind data can be used in weather forecasting. Weather forecasting is the application of science and technology to predict the state of the atmosphere for a given location.Weather warnings are important forecasts because they are used to protect life and property.

![image3](http://104.236.88.249/wp-content/uploads/2016/04/wunderground_weather_chart.png)

## 2.Air quality and pollutions

### The Air Quality Health Index compiled by local department informs the public of the short-term health risk of air pollution and helps the public take precautionary measures to protect their health

![alt text](https://s-media-cache-ak0.pinimg.com/originals/f8/7a/45/f87a45a980007ba21f300954ce214fb7.png "Title")

## 3.River data

### All of life depends on water, and all of us are citizens of a watershed. Our activities and the ways we use our water resources affect the quality of our drinking water. So monitoring the quality and quantity of surface-water resources is very important

```python
import matplotlib.pyplot as plt
pd.options.display.mpl_style = 'default' # import ggplot moduel
df2.plot(kind = 'box')
```

```python
import seaborn as sns       # use seaborn module
plt.show(sns.boxplot(df2,))  # drow a box type plot
```

```python
plt.show(sns.violinplot(df2, widths = 0.5)) # Draw a combination of boxplot and kernel density estimate
```

```python
plt.show(sns.distplot(df2.ix[:,2], rug = True, bins = 15))
# Flexibly plot a univariate distribution of observations,
# the combination of matplotlib hist plot with seaborn kedplot 
# used for normal distribution function
```

```python
with sns.axes_style("white"):
    plt.show(sns.jointplot(df2.ix[:,1], df2.ix[:,2], kind = "kde"))
# replace the scatterplots and histograms with density estimates
# Draw a plot of two variables with bivariate and univariate graphs
```
