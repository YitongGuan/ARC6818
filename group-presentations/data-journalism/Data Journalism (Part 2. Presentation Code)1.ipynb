### A Really Useful Function: <font color=red >urllib.urlretrieve(url, local, callbackfunc)</font>   ---- #*urllib.<font size=5 >request.</font>urlretrieve(url, local, callbackfunc)(Python 3)*

```python
import urllib # urllib.request(Python 3)
def callbackfunc(blocknum, blocksize, totalsize):
    # quantity of data block, size of data block, total size of the entire file
    percent = 100.0 * blocknum * blocksize / totalsize
    if percent > 100:
        percent = 100
    print ("%.2f%%"% percent) # '.2f' means the type of number is float, retaining two digits after the decimal point
url = 'http://www.sina.com.cn'
local = 'd:\\sina.html'
urllib.urlretrieve(url, local, callbackfunc) # urllib.urlretrieve(url, filename=None, reporthook=None, data=None)
```

### What is Principal Component Analysis ?
In simple words, principal component analysis is a method of extracting important variables (in form of components) from a large set of variables available in a data set. It extracts low dimensional set of features from a high dimensional data set with a motive to capture as much information as possible. With fewer variables, visualization also becomes much more meaningful. PCA is more useful when dealing with 3 or higher dimensional data.
It is always performed on a symmetric correlation or covariance matrix. This means the matrix should be numeric and have standardized data.

![](https://www.analyticsvidhya.com/wp-content/uploads/2016/03/2-1-e1458494877196.png)

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2) # in 2 dimensions
pca.fit(existing_df) # process the data using PCA to reduce dimensions

existing_2d = pca.transform(existing_df) # return the data which has been processed by PCA 
existing_df_2d = pd.DataFrame(existing_2d) # put the processed data in DataFrame by Pandas
existing_df_2d.index = existing_df.index # set the heads of rows
existing_df_2d.columns = ['PC1','PC2'] # set the heads of columns

%matplotlib inline # get some function from iPython

ax = existing_df_2d.plot(kind='scatter', x='PC2', y='PC1', figsize=(16,8))  # style setting. x axis, y axis.

for i, country in enumerate(existing_df.index):
    ax.annotate(country, (existing_df_2d.iloc[i].PC2, existing_df_2d.iloc[i].PC1)) # annotate the names of countries; define the x, y
    
#for i in range(len(L)):    # find the length of the list
    #item = L[i]
    ## ... compute some result based on item ...
    #L[i] = result

#This can be rewritten using enumerate() as:

#for i, item in enumerate(L):
     ## ... compute some result based on item ...
     #L[i] = result
```

There are more countries at the very top and on the right hand side. That is, there are more countries with relatively less cases and that have improved their situation with time. But there are some facts that makes us want to keep exploring our data:  
- The distribution is not uniform. There is a high concentration of countries at the very top. This means that there are a group of countries able to deal with the disease in a way that is not accesible to many more (i.e. those countries over and under the 0 line in the vertical axis respectively).  
- There are countries that, far from improving its situation in the 1990-2007 period has most of them have done, ended up with more cases in 2007 than they had in 1990.  
- There is a country, [Djibouti](https://en.wikipedia.org/wiki/Djibouti) that lies quite outside of the rest and might need closer attention.  

## Groping countries based on their situation  
By [clustering](https://en.wikipedia.org/wiki/Cluster_analysis) our original dataset, we will colour the previous chart in order to identify up to six different groups. This is the minimal number of clusters that we have observed better (and more consistently) separates them in order to have a look at the center of each cluster, and try to understand different world tendencies and differences.    

```python
import numpy as np
from sklearn.cluster import KMeans # centralize the similar data in each groups

np.random.seed(1234)  # It will generate the same random number from '1234'
kmeans = KMeans(n_clusters=6) # categrize the data in 6 groups
clusters = kmeans.fit(existing_df) # the process of centralizing
existing_df_2d['cluster'] = pd.Series(clusters.labels_, index=existing_df_2d.index)

import numpy as np
styles = ['b','r','y','c','g','k'] #colors
existing_df_2d.plot(  # process of drawing
    kind='scatter', #styles
    x='PC2',y='PC1',
    c=[styles[i] for i in existing_df_2d.cluster.astype(np.int)], # arrange colors
    s=100, # size
    figsize=(16,8))
```

Most clusters divide the distribution vertically (i.e. based on the absolute number of cases) but there is also a division in the horizontal axis that we will have a look at. In order to complement this view, let's have a look at the same data as a line chart that represents how each cluster representative evolves over time.  

```python
cluster_centers_df = pd.DataFrame( 
    clusters.cluster_centers_, # set the data in lines
    index=['World Cluster 1', 'World Cluster 2', 'World Cluster 3', 'World Cluster 4', 'World Cluster 5', 'World Cluster 6']) # labels
cluster_centers_df.columns = existing_df.columns # years

ax = cluster_centers_df.T.plot(
    figsize=(16,8),
    style=styles,
    linewidth = 3)
ax.set_ylabel("Existing cases per 100K")
```
