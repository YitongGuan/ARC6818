### Example one: Plotting data in basic Python lists as a line chart.

```python
from bokeh.plotting import figure, output_file, show

x = [1, 2, 3, 4, 5]  # create data for yourself
y = [6, 7, 2, 4, 5]

output_file("lines.html")  # output to static HTML file

p = figure(title="simple line example", x_axis_label='x', y_axis_label='y')  # create a new plot with a title and axis labels

p.line(x, y, legend="Temp.", line_width=2)    # add a line renderer with legend and line thickness. The line width values for the annular wedges. (default 1)

show(p)  # show the results
```

*When you execute this script, you will see that a new output file "lines.html" is created*

### Example two: Combine multiple graphs

```python
from bokeh.plotting import figure, output_file, show

x = [0.1, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0] # prepare some data
y0 = [i**2 for i in x]     # define figures of Y respectively
y1 = [10**i for i in x]
y2 = [10**(i**2) for i in x]

output_file("log_lines.html")   # output to static HTML file and create a new website

p = figure(                     # create a new plot(title,label)
   tools="pan,box_zoom,reset,save",
   y_axis_type="log", y_range=[0.001, 10**11], title="log axis example",
   x_axis_label='sections', y_axis_label='particles'
)

p.line(x, x, legend="y=x")     # add some renderers
p.circle(x, x, legend="y=x", fill_color="white", size=8)
p.line(x, y0, legend="y=x^2", line_width=3)
p.line(x, y1, legend="y=10^x", line_color="red")
p.circle(x, y1, legend="y=10^x", fill_color="red", line_color="red", size=6)    # line plus circle
p.line(x, y2, legend="y=10^x^2", line_color="orange", line_dash="4 4")          # dash means tie two points to a line

# show the results
show(p)
```

### Example three: Bar Plot

```python
from bokeh.sampledata.autompg import autompg  # call sampledata from bokeh
from numpy import array   # Two dimensional numpy arrays are indexed using a[i,j]
 
grouped = autompg.groupby("yr")    # define autompg.groupby('yr) as grouped
mpg = grouped["mpg"]      # define grouped['mpg'] as mpg
avg = mpg.mean()          # define mpg.mean as avg
std = mpg.std()           # define mpg.std as std
years = array(list(grouped.groups.keys()))
american = autompg[autompg["origin"]==1]
japanese = autompg[autompg["origin"]==3]
```

```python
p = figure()   # prepare for data creation

p.quad(left=years-0.4, right=years+0.4, bottom=avg-std, top=avg+std, fill_alpha=0.4)  # creat quad

p.circle(x=japanese["yr"], y=japanese["mpg"], size=8,    # define the features of circle
         alpha=0.4, line_color="red", fill_color=None, line_width=2)   # The fill alpha values for the annular wedges. (default 1.0)

p.triangle(x=american["yr"], y=american["mpg"], size=8,    # define the features of triangle
           alpha=0.4, line_color="blue", fill_color=None, line_width=2)

show(p)    # show the results
```

### Example four: Annular Wedges

```python
from math import log, sqrt   # math.sqrt:return the square root of x   # math.log:return the natural logarithm of x 

import numpy as np     # numpy is an extension, multi-dimensional arrays and matrices
import pandas as pd
from six.moves import cStringIO as StringIO  # StringIO is an example of a file-like object that manages its contents in memory instead of on disk


from bokeh.plotting import figure, output_notebook, show

# import figures
antibiotics = """               
bacteria,                        penicillin, streptomycin, neomycin, gram
Mycobacterium tuberculosis,      800,        5,            2,        negative
Salmonella schottmuelleri,       10,         0.8,          0.09,     negative
Proteus vulgaris,                3,          0.1,          0.1,      negative
Klebsiella pneumoniae,           850,        1.2,          1,        negative
Brucella abortus,                1,          2,            0.02,     negative
Pseudomonas aeruginosa,          850,        2,            0.4,      negative
Escherichia coli,                100,        0.4,          0.1,      negative
Salmonella (Eberthella) typhosa, 1,          0.4,          0.008,    negative
Aerobacter aerogenes,            870,        1,            1.6,      negative
Brucella antracis,               0.001,      0.01,         0.007,    positive
Streptococcus fecalis,           1,          1,            0.1,      positive
Staphylococcus aureus,           0.03,       0.03,         0.001,    positive
Staphylococcus albus,            0.007,      0.1,          0.001,    positive
Streptococcus hemolyticus,       0.001,      14,           10,       positive
Streptococcus viridans,          0.005,      10,           40,       positive
Diplococcus pneumoniae,          0.005,      11,           10,       positive
"""

drug_color = {                       # define color of each antibiotics
    "Penicillin"   : "#0d3362",
    "Streptomycin" : "#c64737",
    "Neomycin"     : "black",
}

gram_color = {                       # define the backgroud color of the circle
    "positive" : "#aeaeb8",
    "negative" : "#e69584",
}

df = pd.read_csv(StringIO(antibiotics), skiprows=1, skipinitialspace=True, engine='python')

width = 800     
height = 800
inner_radius = 90    # define radium of inner and outer circles
outer_radius = 300 - 10

minr = sqrt(log(.001 * 1E4))  # 1E4 = 10000  √lg(0.001*10**4)
maxr = sqrt(log(1000 * 1E4))  # √lg(1000*10**4)
a = (outer_radius - inner_radius) / (minr - maxr)
b = inner_radius - a * maxr

big_angle = 2.0 * np.pi / (len(df) + 1)  # len=length   pi=π
small_angle = big_angle / 7

def rad(mic):
    return a * np.sqrt(np.log(mic * 1E4)) + b

output_notebook()  # make sure it could work
```

```python
p = figure(plot_width=width, plot_height=height, toolbar_sticky=False,  # define features of figures
           x_axis_type=None, y_axis_type=None,   # there aren't x_axis and y_axis which area set as none
           x_range=(-420, 420), y_range=(-420, 420), 
           min_border=0, outline_line_color="black",   # define outline and background color
           background_fill_color="#f0e1d2", border_fill_color="#f0e1d2")


angles = np.pi/2 - big_angle/2 - df.index.to_series()*big_angle  # annular wedges
colors = [gram_color[gram] for gram in df.gram]
p.annular_wedge(0, 0, inner_radius, outer_radius, -big_angle+angles, angles, color=colors)

p.annular_wedge(0, 0, inner_radius, rad(df.penicillin), -big_angle+angles + 5*small_angle,  # small wedges 
                -big_angle+angles+6*small_angle, color=drug_color['Penicillin'])
p.annular_wedge(0, 0, inner_radius, rad(df.streptomycin), -big_angle+angles + 3*small_angle, 
                -big_angle+angles+4*small_angle, color=drug_color['Streptomycin'])
p.annular_wedge(0, 0, inner_radius, rad(df.neomycin), -big_angle+angles + 1*small_angle, 
                -big_angle+angles+2*small_angle, color=drug_color['Neomycin'])


labels = np.power(10.0, np.arange(-3, 4))  # circular axes and lables
radii = a * np.sqrt(np.log(labels * 1E4)) + b
p.circle(0, 0, radius=radii, fill_color=None, line_color="white")
p.text(0, radii[:-1], [str(r) for r in labels[:-1]], 
       text_font_size="8pt", text_align="center", text_baseline="middle")


p.annular_wedge(0, 0, inner_radius-10, outer_radius+10,   # black radial axes
                -big_angle+angles, -big_angle+angles, color="black")


xr = radii[0]*np.cos(np.array(-big_angle/2 + angles))  # bacteria labels
yr = radii[0]*np.sin(np.array(-big_angle/2 + angles))
label_angle=np.array(-big_angle/2+angles)
label_angle[label_angle < -np.pi/2] += np.pi # easier to read labels on the left side
p.text(xr, yr, df.bacteria, angle=label_angle, text_font_size="9pt", text_align="center", text_baseline="middle")


p.circle([-40, -40], [-370, -390], color=list(gram_color.values()), radius=5)  # these hand drawn legends are pretty clunky,which will be improved in future release
p.text([-30, -30], [-370, -390], text=["Gram-" + x for x in gram_color.keys()], 
       text_font_size="7pt", text_align="left", text_baseline="middle")

p.rect([-40, -40, -40], [18, 0, -18], width=30, height=13, color=list(drug_color.values()))
p.text([-15, -15, -15], [18, 0, -18], text=list(drug_color.keys()),
       text_font_size="9pt", text_align="left", text_baseline="middle")

p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

show(p)
```

### Example five: Periodic-table of elements

```python
from bokeh.models import HoverTool, ColumnDataSource
from bokeh.plotting import figure, show
from bokeh.sampledata.periodic_table import elements    # import sampledata

romans = ["I", "II", "III", "IV", "V", "VI", "VII"]     # denfine romans figures

elements["atomic mass"] = elements["atomic mass"].astype(str)

elements["period"] = [romans[x-1] for x in elements.period]
elements = elements[elements.group != "-"]

group_range = [str(x) for x in range(1, 19)]

colormap = {                                # denine colors for each sort of elements.
    "alkali metal"         : "#a6cee3",
    "alkaline earth metal" : "#1f78b4",
    "halogen"              : "#fdbf6f",
    "metal"                : "#b2df8a",
    "metalloid"            : "#33a02c",
    "noble gas"            : "#bbbb88",
    "nonmetal"             : "#baa2a6",
    "transition metal"     : "#e08e79",
}

source = ColumnDataSource(     # classify elements 
    data=dict(
        group=[str(x) for x in elements["group"]],
        period=[str(y) for y in elements["period"]],
        symx=[str(x)+":0.1" for x in elements["group"]],
        numbery=[str(x)+":0.8" for x in elements["period"]],
        massy=[str(x)+":0.15" for x in elements["period"]],
        namey=[str(x)+":0.3" for x in elements["period"]],
        sym=elements["symbol"],
        name=elements["name"],
        cpk=elements["CPK"],
        atomic_number=elements["atomic number"],
        electronic=elements["electronic configuration"],
        mass=elements["atomic mass"],
        type=elements["metal"],
        type_color=[colormap[x] for x in elements["metal"]],
    )
)
 
p = figure(title="Periodic Table", tools="hover,save",                  
           x_range=group_range, y_range=list(reversed(romans)))     # set title, tools  of figure
p.plot_width = 1200
p.toolbar_location = None
p.outline_line_color = None

p.rect("group", "period", 0.9, 0.9, source=source,    # draw rectangle of each elements
       fill_alpha=0.6, color="type_color")

text_props = {
    "source": source,
    "angle": 0,
    "color": "black",
    "text_align": "left",
    "text_baseline": "middle"
}

p.text(x="symx", y="period", text="sym",        # edit information for every elements in different groups.
       text_font_style="bold", text_font_size="15pt", **text_props)

p.text(x="symx", y="numbery", text="atomic_number",    
       text_font_size="9pt", **text_props)

p.text(x="symx", y="namey", text="name",
       text_font_size="6pt", **text_props)

p.text(x="symx", y="massy", text="mass",
       text_font_size="5pt", **text_props)

p.grid.grid_line_color = None

p.select_one(HoverTool).tooltips = [        # classification
    ("name", "@name"),
    ("atomic number", "@atomic_number"),
    ("type", "@type"),
    ("atomic mass", "@mass"),
    ("CPK color", "$color[hex, swatch]:cpk"),
    ("electronic configuration", "@electronic"),
]

show(p)  # Change to save(p) to save but not show the HTML file
```
