##                              Some Useful Settings and Methods for Turtle
------------------------------------------------------------
            These tables show some of the values for various methods used in the Python turtle package.
### **Speed**
The methods that require a speed with which to move the cursor (turtle) take an integer argument. The integer is between 0 and 10, with 0 meaning no animated motion and 10 meaning the fastest animated motion. So if you have the speed set to 0, the turtle will jump around as you move it; there will be no animation.
>**Turtle Commands**

>• `tur.speed(n )` sets the speed of the turtle tur to n, where n is as above.

>• `turtle.tracer(num,num)` By calling tracer() with bother arguments set to zero, we are essentially turning off all
animation and our drawings will be drawn “immediately.” However, if we turn the animation off in this way we need to explicitly update the image with the `update()` method after we are done issuing drawing commands. If you want to reset tracer() to its original settings, its arguments should be 1 and 10.

### **Color**
The methods that require a color use a color specification string. The more common ones are:
>`black` | `cyan` | `blue` | `gold` | `brown` | `gray`|
>`green` | `magenta` | `orange`|
>`pink` | `violet` | `purple`  |`white` | `red` | `yellow`

A full list can be found at http://www.tcl.tk/man/tcl8.5/TkCmd/colors.htm. You can also specify colors as mixtures or red, green, and blue as #rrggbb , where rr, gg, and bb are two hexadecimal digits each indicating the intensity of the color (with “00” meaning the color isn’t present and “FF” meaning it is as intense as possible).

>**Turtle Commands**

>• `win.bgcolor(col )` sets the background color of window win to col, where col is as above.

>• `tur.pencolor(col)` sets the color of the pen with which turtle draws on the screen.

### **Shapes**
The methods that require a shape for the cursor (turtle) know the following shapes: 
>`arrow` | `turtle` | `circle` | `square` | `triangle` | `classic`

>**Turtle Commands**

>• `tur.shape(img )` sets the shape of the turtle tur to img, where img is as above.

>• `tur.stamp()` stamps an image of the turtle cursor shape onto the screen. Alternately `tur.clearstamp()` clears all stamps of the specific turtle from the screen.

### **Drawing Things**
A “pen” is associated with each turtle. The pen can be in the “up” state or the “down” state. When you move the cursor (turtle), if the pen is down, the turtle draws a line from the current position to the new one. If the pen is up, the position of the turtle changes to the new one, but no line is drawn.
You can draw a dot or a circle of a given size, too.
You can also write text at the current location of the mouse.
>**Turtle Commands**

>• `tur.goto(x, y)` moves the turtle tur from the current co-ordinates to (x, y). Alternatively, **tur.setx(int)** and **tur.sety(int)** methods work to set only the x or y coordinates of the turtle (where int is a numerical coorndinate). `tur.xcor()` and `tur.ycor()` return the current coordinates of the x or y state of the turtle 'tur'.

>• `tur.penup()` raises the pen (puts it into the “up” state) associated with the turtle tur. When you
move the turtle, the co-ordinates change but no line is drawn.

>• `tur.pendown()` lowers the pen (puts it into the “down” state) associated with the turtle tur. When
you move the turtle, the co-ordinates change and a line is drawn from the current co-ordinates to the
given co-ordinates.

>• `turtle.pensize(width)` or `turtle.width(width)` Set the line thickness to width or return it. If no argument is given, the current pensize is returned. Parameters:	width – a positive number. Alternately, `turtle.pen()` returns or set the pen’s attributes as in the following example: 

>turtle.pen(fillcolor="black", pencolor="red", pensize=10)


>• `tur.dot(n )` draws a dot at the current position of the turtle tur. If n is present, the dot has diameter
n pixels. If n is not present, the dot has a default size (the maximum of pensize + 4 and 2×pensize, where pensize is the width used to draw a line, if you must know).

>• `tur.circle(n )` draws a circle of radius n pixels, with the center n pixels left of the current position of the turtle tur.

>• `tur.write(string, move, align, (fontname, fontsize, fonttype ))` writes the given string at the current position of the turtle tur. It is positioned as indicated by align, which may take the (string) values "left", "center", or "right". The text is printed in the named fontname, with point size fontsize, in the type fonttype. Then, if move is True, the position of the turtle tur changes to the bottom right corner of the text. If move is omitted, it defaults to False; if align is omitted, it defaults to "left"; and if the font triplet is omitted, the font Ariel, with point size 8, in normal type, is used.

>• `tur.hideturtle()` hides the turtle 'tur' cursor. Alternately `tur.showturtle()` makes the turtle cursor visible. 

### **Coloring in Shapes (Filling Shapes)**
You can color in, or “fill”, a shape such as a rectangle or circle (or, really, anything). To do this, say you want to begin filling the shape then draw the shape, then say you are done. Filling is associated with a turtle, so if you have several active turtles, the filling is done only for the one(s) you indicate.
By default, the fill color is the color of the turtle. You can change it to another color as described in the section Color, above.
>**Turtle Commands**

>• `tur.begin_fill()` says to begin filling the shape subsequently drawn using turtle tur.

>• `tur.end_fill()` says to stop filling the shape drawn using turtle tur.

>• `tur.fillcolor(col)` sets the fill color associated with the turtle tur to col, where col is as in the
Color section above. 

### **Window Control**
Most of the examples in this section refer to a TurtleScreen instance called screen. 
>**Turtle Commands**

>•`turtle.bgcolor(*args)` Set or return background color of the TurtleScreen.
_Parameters:_ args – a color string or three numbers in the range 0...255 or hex value of a color.

>•`turtle.bgpic(picname)` Set background image or return name of current backgroundimage. If picname is a filename, set the corresponding image as background. If picname is "nopic", delete background image, if present. If picname is None, return the filename of the current background image.
_Parameters:_	picname – a string, name of a gif-file or "nopic", or None

>•`turtle.clear()` or `turtle.clearscreen()` Delete all drawings and all turtles from the TurtleScreen. Reset the now empty TurtleScreen to its initial state: white background, no background image, no event bindings and tracing on.

>•`turtle.reset()` or `turtle.resetscreen()` Reset all Turtles on the Screen to their initial state.

>•`turtle.screensize(canvwidth, canvheight, bg)` Sets display screen size and background.
_Parameters:_	_canvwidth_ – positive integer, new width of canvas in pixels, _canvheight_ – positive integer, new height of canvas in pixels, _bg_ – colorstring or color code, new background color

>•`turtle.setworldcoordinates(llx, lly, urx, ury)`

>_Parameters:_	

>llx – a number, x-coordinate of lower left corner of canvas

>lly – a number, y-coordinate of lower left corner of canvas

>urx – a number, x-coordinate of upper right corner of canvas

>ury – a number, y-coordinate of upper right corner of canvas

>• `turtle.title(titlestring)`Set title of turtle window to the contents of 'titlestring'. Parameters:	titlestring – a string that is shown in the titlebar of the turtle graphics window

>•`turtle.bye()`Shuts the turtlegraphics window.

>•`turtle.exitonclick()` Binds bye() method to mouse clicks on the Screen.

## Here's a short story about Joe and Jenny created using turtles
[![JOE & JENNY](http://i.imgur.com/pxNY1pS.png)](https://www.youtube.com/watch?v=t5gLmOzCmdU "JOE & JENNY" )

```python
## Understanding the methods mentioned above through Joe and Jenny's story

import turtle
## Setting up turtle's display window
turtle.setup(1000, 600)
turtle.screensize(1000, 600)                ## Defining size of display screen.
turtle.setworldcoordinates(0, 0, 1000, 600) ## 0,0     = lower left coordinates of display screen
                                            ## 1000,600= upper right coordinates of display screen
turtle.bgcolor('light blue')                ## Setting light blue as background colour for display screen.
turtle.title("Joe and Jenny's Story")       ## Setting title for display window.
```

```python
## Defining Joe--the hero of our story!
joe=turtle.Turtle()
joe.penup()                                 ## Pulling the pen off the canvas
joe.setpos(200,300)                         ## Positioning Joe over the left half of the screen
joe.shape('circle')                         ## Changing turtle shape to circle

# Building Joe's head
joe.color('white')                          ## Setting the pen's colour to white
joe.pendown()                               ## Restoring the pen down on the canvas
joe.begin_fill()                            ## Giving the command to begin colouring the following shape
joe.circle(100)                             ## Constructing a circle with radius=100 
joe.end_fill()                              ## Wrapping up the fill command

# Drawing Joe's neck
joe.seth(-90)                               ## Set the heading of the turtle cursor to -90, i.e. turning it... 
joe.fd(25)                                  ## ...perpendicular in a clockwise fashion (setheading() or seth())

# Joe needs a pair of hands!
joe.left(90)
joe.fd(50)
joe.up()
joe.backward(100)
joe.down()                                  ## down() works synonomous to pendown()
joe.fd(50)

# Defining Joe's 6-pack torso
joe.rt(90)
joe.fd(100)

# Joe has really long feet!
joe.left(90)
joe.fd(60)
joe.up()
joe.backward(120)
joe.down()
joe.fd(60)
joe.color('black')

# One set of dreamy eyes coming up
joe.left(90)
joe.up()
joe.fd(225)
#left eye
joe.left(90)
joe.fd(50)
joe.stamp()
#right eye
joe.back(100)
joe.stamp()                                 ## Imprinting the circle of the turtle on the screen to use as joe's eyes
joe.hideturtle()                            ## Hiding turtle cursor 

```

```python
# The quintessential beauty-^drumroll^-presenting Jenny!
jenny=turtle.Turtle()
jenny.up()                                  ## up() synonomous with penup()
jenny.setpos(800,300)                       ## Setting Jenny over at the right half of the screen
jenny.shape('circle')
jenny.color('white')
jenny.speed(0)                              ## Setting speed of animation to the fastest setting-0
jenny.down()

# Drawing Jenny's head
jenny.begin_fill()                          ## Begin fill operation
jenny.circle(100)
jenny.end_fill()                            ## End colour fill operation

# Jenny has a neck as graceful as that of a swan's
jenny.seth(-90)
jenny.fd(25)

# Building Jenny's hands
jenny.left(90)
jenny.fd(50)
jenny.up()
jenny.backward(100)
jenny.down()
jenny.fd(50)

#Jenny's petite torso
jenny.rt(60)
for i in range(3):                           ## Decking Jenny in a frilly skirt through use of the for loop
    jenny.fd(100)                            ## For loop creates the equilateral triangle for jenny's skirt
    jenny.rt(120)

jenny.color('black')

# Jenny's jet black eyes
jenny.seth(90)
jenny.up()
jenny.fd(100)
jenny.left(90)
jenny.fd(50)
jenny.stamp()
jenny.back(100)
jenny.stamp()                                 ## Stamp() stamps its image on the screen leaving behind Jenny's eyes

jenny.hideturtle()
```

```python
#It's love at first sight
heart=turtle.Turtle()
heart.penup()
heart.setpos(500,300)
heart.down()
heart.color('pink','red')                     ## Setting pencolor as pink and fill color as red.
heart.begin_fill()
## Making the heart shape
heart.left(45)                                
heart.forward(90)
heart.circle(45, 180)
heart.right(90)
heart.circle(45, 180)
heart.forward(90)
heart.end_fill()
heart.hideturtle()
```

```python
## And the rest, as they say, is history
text=turtle.Turtle()
text.up()
text.hideturtle()
text.color('black')
text.setpos(250,200)
text.write('...And they lived happily ever after', font=('helvetica', 30, 'bold')) ## write() to display text on screen
```

```python
heart.reset()                                 ## Reseting the turtles joe, jenny and heart, i.e. they... 
joe.reset()                                   ## ...disappear/revert back to their original state
jenny.reset()

# Living happily in not-so-sunny Sheffield
cloud=turtle.Turtle()
cloud.speed(0)
cloud.color('white')
cloud.up()
cloud.setpos(50,450)
cloud.hideturtle()
for i in range(3):                            ## Setting up for loop to create clouds
    cloud.setpos(cloud.xcor()+200,cloud.ycor()) ## cloud.xcor() returns the x current coordinate of the turtle 'cloud'..
    cloud.down()                                ## ..cloud.ycor() returns the y current coordinate of the turtle 'cloud'
    cloud.fd(100)
    cloud.begin_fill()
    cloud.circle(45, 180)
    cloud.rt(75)
    cloud.circle(45, 180)
    cloud.rt(75)
    cloud.circle(45, 100)
    cloud.end_fill()
    cloud.left(50)
    cloud.up()                              ## That's it folks! We just finished telling our first story using turtles!
```


### Turtles in NetLogo
> A second programming language was developed from Logo, NetLogo. 
It is an _agent-based model_ , which means that it simulates the action and interaction of  agents and provides an outcome of the whole group. It is used to simulate hypotethesis and predict patterns of behaviour.

>NetLogo is composed by _agents_ - elements that follow commands.
There are 4 types of agents: 
#### Patches 
> The 2D world of NetLogo is divide by a grid, each square formed by the grid is a Patch. The patches compose the ground where Turtles walk upon.
#### Turtles
> Agent that moves around the Patches
#### Link
> The conection between two turtles
#### Observer
>With no material representation, it is whoever inputs commands to the other agent

>**Event- Driven Programming**

>• There are several ways in which the user can give input for the turtles, a few of the methods are illustrated below:

>' Keypress events '

>• Each keypress can be configured to achieve a certain purpose: In relation to games it would essentially be controlling the main character's (turtle) movements so, the pressing of the LEFT, RIGHT, UP or DOWN keys would initiate different functions.

   The other elements (monsters!) in the game can be equated to other turtles whose movement is in relation to the main character


>' Mouse events '

>• These are different to key press events and could be handled in two ways:

> A simple click of a particular mouse button with out the actual need to locate the cursor. This could for example simply start or end a program.

> Another method of input would be to identify the exact position of the mouse on the screen and initiate a function in relation to that specific input. This therefore, requires the X and Y co-ordinates of the position of the cursor when the mouse was clicked.

> Further, individual turtles can also have their own handlers for mouse clicks, thereby exibiting different attributes on being clicked.

>' Automatic Events Based on Timer'

>• This is merely a precursor to the next group's presentation on Python's time module.

>• It is also possible for turtles to attain certain sttributes or perform certain functions after a certain fixed intervals of time creating automatic events.

>• Python programming has been utilised in traffic signals : so, the best example to visualize the above event- driven programming would be to envision the Sheffield traffic system as the model

>> The pedestrian pushes the button at a crossing:

>> After a particular amount of time the pedestrian signal changes to green.

>> This lasts for a fixed interval of time after which the signal changes to orange and successively to red.

>> The traffic signal is the turtle whose state changes based on the timer which in turn is initiated by the input from the user.
