## Â·Create wordcloud with your own image and text data.
### Before we do this exercise, we should install [WordCloud](https://github.com/amueller/word_cloud).
### Notice: Maybe you need to install [C++ for 3.5](http://landinghub.visualstudio.com/visual-cpp-build-tools) or [C++ for 2.7](https://www.microsoft.com/en-us/download/confirmation.aspx?id=44266) in your computer before install WorldCloud

```python
from os import path
from scipy.misc import imread
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
d = path.dirname('.')
text = open(path.join(d, "Harry Potter and the Sorcerer's Stone.txt")).read()
Harry_coloring = imread(path.join(d, "Harry Potter and the Sorcerer's Stone.png"))
```

```python
wc = WordCloud(background_color="white",#set the backgroud color you like 
max_words=2000,                         #set the maximum number of words that you want to show your wordcloud
mask=Harry_coloring,                    #set the backgroup iamge
stopwords = set(STOPWORDS),  
max_font_size=40,                       #set the maximum size of word show in word cloud
random_state=42)

wc.generate(text)                       #create word cloud

image_colors = ImageColorGenerator(Harry_coloring)  #create colors from backgroud image 
```

```python
plt.imshow(wc)
plt.axis("off")

plt.figure()
# recolor wordcloud and show
# we could also give color_func=image_colors directly in the constructor
plt.imshow(wc.recolor(color_func=image_colors))
plt.axis("off")
# draw wordcloud which follow image's color
plt.figure()
plt.imshow(Harry_coloring, cmap=plt.cm.gray)
plt.axis("off")
plt.show()
# save image
wc.to_file(path.join(d, "Harry.png"))

## reference: https://github.com/amueller/word_cloud
```

# Example 2. Catch twitter user data base on Tweepy
## All code and comments below from [Adil Moujahid](http://adilmoujahid.com/posts/2014/07/twitter-analytics/)

## We will be using a Python library called Tweepy to connect to Twitter Streaming API and downloading the data. If you don't have Tweepy installed in your machine, go to this [link](https://github.com/tweepy/tweepy), and follow the installation instructions

```python
#Import the necessary methods from tweepy library
from tweepy.streaming import StreamListener
from tweepy import OAuthHandler
from tweepy import Stream

#Variables that contains the user credentials to access Twitter API 
#Make sure to enter your credentials into access_token, access_token_secret, consumer_key, and consumer_secret.
access_token = "ENTER YOUR ACCESS TOKEN"
access_token_secret = "ENTER YOUR ACCESS TOKEN SECRET"
consumer_key = "ENTER YOUR API KEY"
consumer_secret = "ENTER YOUR API SECRET"


#This is a basic listener that just prints received tweets to stdout.
class StdOutListener(StreamListener):

    def on_data(self, data):
        print(data)
        return True

    def on_error(self, status):
        print(status)


if __name__ == '__main__':

    #This handles Twitter authetification and the connection to Twitter Streaming API
    l = StdOutListener()
    auth = OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_token_secret)
    stream = Stream(auth, l)

    #This line filter Twitter Streams to capture data by the keywords: 'python', 'javascript', 'ruby'
    stream.filter(track=['python', 'javascript', 'ruby'])
```

## The data that we stored twitter_data.txt is in JSON format

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the data in into an array that we call tweets.

tweets_data_path = 'twitter_data.txt'

tweets_data = []
tweets_file = open(tweets_data_path, "r")
for line in tweets_file:
    try:
        tweet = json.loads(line)
        tweets_data.append(tweet)
    except:
        continue
```

```python
len(tweets_data)
```

```python
# structure the tweets data into a pandas DataFrame to simplify the data manipulation

tweets = pd.DataFrame()
```

```python
# add 3 columns to the tweets DataFrame called text, lang, and country. 
# text column contains the tweet,
# lang column contains the language in which the tweet was written, 
# and country the country from which the tweet was sent.

tweets['text'] = map(lambda tweet: tweet['text'], tweets_data)
tweets['lang'] = map(lambda tweet: tweet['lang'], tweets_data)
tweets['country'] = map(lambda tweet: tweet['place']['country'] if tweet['place'] != None else None, tweets_data)
tweets
```

```python
#The first one describing the Top 5 languages in which the tweets were written

tweets_by_lang = tweets['lang'].value_counts()

fig, ax = plt.subplots()
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=10)
ax.set_xlabel('Languages', fontsize=15)
ax.set_ylabel('Number of tweets' , fontsize=15)
ax.set_title('Top 5 languages', fontsize=15, fontweight='bold')
tweets_by_lang[:5].plot(ax=ax, kind='bar', color='red')
```

```python
tweets_by_country = tweets['country'].value_counts()

fig, ax = plt.subplots()
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=10)
ax.set_xlabel('Countries', fontsize=15)
ax.set_ylabel('Number of tweets' , fontsize=15)
ax.set_title('Top 5 countries', fontsize=15, fontweight='bold')
tweets_by_country[:5].plot(ax=ax, kind='bar', color='blue')
```
