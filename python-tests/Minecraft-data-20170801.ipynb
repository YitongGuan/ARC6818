### Analyse Minecraft log data with python 

##### create a new text file with a comma at the end of each line, making this a list of dictionaries

```python
myCharacter = ','
with open('log-new.txt', 'w') as out_file:
    with open('log.txt', 'r') as in_file:
        for line in in_file:
            out_file.write(line.rstrip('\n') + myCharacter + '\n')
```

##### read the the new file in to Python and convert to a pandas dataframe

```python
inf = open('log-new.txt','r')
dict_from_file1 = eval(inf.read())
inf.close()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(dict_from_file1)
```
